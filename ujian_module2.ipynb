## 1. Show a list of 10 films with the shortest duration. Sort data by movies with the shortest duration. Columns that must be displayed are title, category, and length.

```python
query(
    '''
    select title, category, length 
from film_list
where category = 'comedy'
order by length asc limit 10
;
    '''
)
```

## 2. Show a complete list of movie categories along with the number of movies in each category & the average DVD movie rental price for each category. Sort the data from the category with the most number of films. The minimum required fields are category, number of films and average rental price.

```python
query(
    '''
    select category AS movie_category, count(category) AS number_of_movies , avg(price) AS average_rent_price 
from film_list
group by movie_category
order by number_of_movies desc;

    '''
)
```

## 3. Show a list of the top 10 actors/actresses who have starred in the most movies. Minimum displayed column: actor id, first name, last name and number of films starring then sort from actors/actresses who starred in the most films.

```python
query(
    '''

    select fa.actor_id AS actor_id, a.first_name AS first_name, a.last_name AS last_name, count(fa.actor_id)
AS number_of_films from actor a
join film_actor fa on a.actor_id = fa.actor_id
group by actor_id
order by number_of_films desc
Limit 10 ;

    '''
)
```

## 4. From the previous question, it is known that Gina Degeneres is the actress who has starred in the most films, with a total of 42 film titles. What category of films has he starred in the most? To find out, show a list of film categories along with the number of films that Gina Degeneres has starred in. The required fields are the category of film and the number of films starring.

```python
query(
    '''
select fl.category, count(fl.category) AS numbers_of_movie 
from actor a
join film_actor fa on a.actor_id = fa.actor_id
join film f on fa.film_id = f.film_id
join film_list fl on f.title = fl.title
where a.actor_id in (107)
group by fl.category
order by numbers_of_movie desc
;
    '''
)
```

## 5. From the previous question, it is known that Gina Degeneres has starred in the most science fiction genre films, with a total of 7 film titles. Show a list of sci-fi movies that have starred Gina Degeneres. The required fields are the title of the film and its category.

```python
query(
    '''
    select f.title, c.name category
from actor a
join film_actor fa on a.actor_id = fa.actor_id
join film f on fa.film_id = f.film_id
join film_category fc on f.film_id = fc.film_id
join category c on fc.category_id = c.category_id
where a.actor_id in (107) and c.name = 'Sci-Fi'
;
    '''
)
```
