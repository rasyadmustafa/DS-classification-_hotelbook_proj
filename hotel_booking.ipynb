# **Case Study** 
In the hotel business, ensuring the availability of hotel rooms at a time and for some time to come is very important in optimizing revenue. Accurate and actual room information can prevent the hotel from various losses, such as losses caused by guests who make orders or do not check on arrival.

Guests staying overnight, either after informing the hotel in advance (eg contacting themselves or being contacted by the hotel) or not, cause the rooms they should be staying in. These rooms can actually still be rented back to other guests, but it is more difficult to decide which ones can be rented back if the news of the guest's arrival is received by the hotel, for example, sometime before check-in time.

If the room has already been booked for these guests, while they have not checked in, the room may only be rented out the next day. Not to mention if the hotel operator implements a free cancellation policy, which means there is a possibility that the hotel will still have to refund the guest's booking fee. As a result, the hotel suffers revenue losses in the form of opportunity costs for rooms that are not accommodated.

# **Problem Framing and Task**
- ***Business objective*** 
    - Based on the case study in this project, we were asked to create a classification model to predict whether a customer will cancel a hotel booking or  not. This greatly affects the hotel business so as not to suffer losses. 

- ***Output***  
    - By making a predictive model of whether customers who have booked a room will cancel their order or not, based on variables or features in the data we have. The expected benefit for users is that they can optimize hotel business activities, and can reduce losses due to room cancellations so that 
    rooms cannot be rented out again for some time.
    
        NOTE: Our model should only be used for benchmarks purposes; do not depends on the outcome of our model.

- ***Machine learning task*** 
    - By looking at the problem and the expected output based on the problem, the machine learning model that will be used is supervised learning with the classification method because we know the label/target we want to predict, which will cancel the room reservation or not.

- ***Performance measure*** 
    - Because the model to be made is supervised learning (classification), the Performance measure used is Accuracy, which is the ratio of correct predictions (positive and negative) to the overall data. precision Is the ratio of positive correct predictions compared to the overall positive predicted results. Recall is the ratio of true positive predictions compared to the total number of true positive data. Recall answers the question "What percentage of customers are predicted to cancel compared to all customers who actually cancel". and the F1 score which is a weighted comparison of the average precision and recall. However, in this project, the focus will be on **precision** and **accuracy** because the predicted customer will not cancel but actually cancels so that the hotel can take immediate action.

        source reference : https://rey1024.medium.com/mengenal-accuracy-precission-recall-dan-specificity-serta-yang-diprioritaskan-b79ff4d77de8

- ***The Risk*** 
    - which may be caused by prediction errors from the model later, among others, where the customer is predicted to cancel the hotel but actually it is not the other way around. Therefore, it results in an income loss because the room is canceled by an order or does not stay overnight resulting in the room being empty but cannot be rented out to other customers which results in the loss of business opportunities, not to mention the hotel operator implementing a free-cancellation policy, such a risk will be experienced.

# **Features Information**

-   **ADR** :    Average Daily Rate as defined</p> 
-   **Adults** : Number of adults</p>
-   **Agent** : ID of the travel agency that made the booking </p>
-   **ArrivalDateDayOfMonth** : Day of the month of the arrival date </p>
-   **ArrivalDateMonth** : Month of arrival date with 12 categories: “January” to “December” </p>
-   **ArrivalDateWeekNumber** : Week number of the arrival date </p>
-   **ArrivalDateYear** : Year of arrival date </p>
-   **AssignedRoomType** : Code for the type of room assigned to the booking. Sometimes the assigned room type differs from the reserved room type due to   hotel operation reasons (e.g. overbooking) or by customer request. Code is presented instead of designation for anonymity reasons </p>
-   **Babies** : 	Number of babies </p>
-   **BookingChanges** : Number of changes/amendments made to the booking from the moment the booking was entered on the PMS until the moment of check-in or cancellation </p>
-   **Children** : Number of children </p>
-   **Company** : ID of the company/entity that made the booking or responsible for paying the booking. ID is presented instead of designation for anonymity reasons </p>
- **Country** : Country of origin. Categories are represented in the ISO 3155–3:2013 format  </p>
- **CustomerType** : Type of booking, assuming one of four categories :
    - Contract - when the booking has an allotment or other type of contract associated to it
    - Group – when the booking is associated to a group
    - Transient – when the booking is not part of a group or contract, and is not associated to other transient booking
    - Transient-party – when the booking is transient, but is associated to at least other transient booking
    </p>
-   **DaysInWaitingList** : Number of days the booking was in the waiting list before it was confirmed to the customer </p>
-   **DepositType** : Indication on if the customer made a deposit to guarantee the booking. This variable can assume three categories:
    - No Deposit – no deposit was made
    - Non Refund – a deposit was made in the value of the total stay cost
    - Refundable – a deposit was made with a value under the total cost of stay.
    </p>
-   **DistributionChannel** : Booking distribution channel. The term “TA” means “Travel Agents” and “TO” means “Tour Operators” </p>
-   **IsCanceled** : Value indicating if the booking was canceled (1) or not (0), **this is our target variable** </p>
-   **IsRepeatedGuest** : Value indicating if the booking name was from a repeated guest (1) or not (0) </p>
-   **LeadTime** : Number of days that elapsed between the entering date of the booking into the PMS and the arrival date </p>
-   **MarketSegment** : Market segment designation. In categories, the term “TA” means “Travel Agents” and “TO” means “Tour Operators” </P>
-   **Meal** : Type of meal booked. Categories are presented in standard hospitality meal packages:
    - Undefined/SC – no meal package
    - BB – Bed & Breakfast
    - HB – Half board (breakfast and one other meal – usually dinner)
    - FB – Full board (breakfast, lunch and dinner)
    </p>
-   **PreviousBookingsNotCanceled** : Number of previous bookings not cancelled by the customer prior to the current booking </p>
-   **PreviousCancellations** : Number of previous bookings that were cancelled by the customer prior to the current booking </p>
-   **RequiredCardParkingSpaces** : Number of car parking spaces required by the customer </p>
-   **ReservationStatus** : Reservation last status, assuming one of three categories:
    - Canceled – booking was canceled by the customer
    - Check-Out – customer has checked in but already departed
    - No-Show – customer did not check-in and did inform the hotel of the reason why
    </p>
-   **ReservationStatusDate** : Date at which the last status was set. This variable can be used in conjunction with the ReservationStatus to understand when was the booking canceled or when did the customer checked-out of the hotel </p>
-   **ReservedRoomType** : Code of room type reserved. Code is presented instead of designation for anonymity reasons </p>
-   **StaysInWeekendNights** : Number of weekend nights (Saturday or Sunday) the guest stayed or booked to stay at the hotel </p>
-   **StaysInWeekNights** : Number of week nights (Monday to Friday) the guest stayed or booked to stay at the hotel </p>
-   **TotalOfSpecialRequests** : Number of special requests made by the customer (e.g. twin bed or high floor)</p>

# **Data Preparation**

### **Feature Engineering and Preprocessing**

- Note: for features that are not taken in the formation of the model are:
     - **country** : because there are so many unique values from country, it will not be effective when encoding
     - **agent** and **company** : multi-collinearity between the two and not related to the target variable
     - **reservation status date** : because there are so many unique values and the feature should be of the time series type but in this case it is not needed
     - **reservation status** : this feature is always the same value as the target and should only be filled after the label is filled

```python
# Splitting the Data

from sklearn.model_selection import train_test_split

x = df_hotel.drop(['is_canceled','country','agent','company','reservation_status_date','reservation_status'], axis=1)
y = df_hotel['is_canceled']

X_train,X_test,y_train,y_test = train_test_split(x,y, test_size=0.2,stratify=y,random_state=22)
```

```python
x.describe(exclude='number').columns
```

```python
x.describe().columns
```

```python
# Preprocess
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import RobustScaler, StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# define each columns
robust_cols = ['lead_time', 'stays_in_weekend_nights', 'stays_in_week_nights', 'adults', 'children', 'babies', 'previous_cancellations', 'previous_bookings_not_canceled', 'booking_changes', 'days_in_waiting_list', 'adr', 'required_car_parking_spaces']

standard_cols = ['arrival_date_year', 'arrival_date_week_number', 'arrival_date_day_of_month', 'total_of_special_requests']

onehot_cols = ['hotel', 'arrival_date_month', 'meal', 'market_segment', 'distribution_channel', 'reserved_room_type', 'assigned_room_type', 'deposit_type', 'customer_type']

# pipeline for imputer + onehot encoder
simp_imp_robust_pipeline = Pipeline([
    ('simp_imp', SimpleImputer(strategy='median')),
     ('robust_scaler', RobustScaler())
])


# pipeline for not tree based
nt_pipeline = ColumnTransformer([
    ('robust_scaler', simp_imp_robust_pipeline, robust_cols),
    ('standard_scaler', StandardScaler(), standard_cols),
    ('onehot_encoder', OneHotEncoder(handle_unknown='ignore'), onehot_cols)
])

# Pipeline for tree-based model ( No scaler for interpretation purposes)
tb_pipeline = ColumnTransformer([
    ('onehot_encoder', OneHotEncoder(handle_unknown='ignore'), onehot_cols),
    ('simp_imp', SimpleImputer(strategy='median'),robust_cols)
],remainder='passthrough')

X_train_fe_prep = pd.DataFrame(nt_pipeline.fit_transform(X_train))
X_test_fe_prep = pd.DataFrame(nt_pipeline.transform(X_test))

X_train_fe_prep_tree = pd.DataFrame(tb_pipeline.fit_transform(X_train))
X_test_fe_prep_tree = pd.DataFrame(tb_pipeline.transform(X_test))
```

```python
# Get features names (For Feature Importance Purposes)

num_cols = ['lead_time', 'arrival_date_year', 'arrival_date_week_number',
       'arrival_date_day_of_month', 'stays_in_weekend_nights',
       'stays_in_week_nights', 'adults', 'children', 'babies',
       'is_repeated_guest', 'previous_cancellations',
       'previous_bookings_not_canceled', 'booking_changes',
       'days_in_waiting_list', 'adr', 'required_car_parking_spaces',
       'total_of_special_requests']

features_nt = robust_cols + standard_cols + list(nt_pipeline.transformers_[2][1].get_feature_names())
features_tb = list(tb_pipeline.transformers_[0][1].get_feature_names()) + num_cols

# features name not tree
X_train_fe_prep.columns = features_nt
X_test_fe_prep.columns = features_nt

# features name tree based
X_train_fe_prep_tree.columns = features_tb
X_test_fe_prep_tree.columns = features_tb
```

# **MODELING**

- For our project we will use 4 models:
    - Logistic Regression
    - KNN
    - RandomForest
    - XGBoost

- These 4 models will be trained as is then we will also perform hyperparameter tuning for each models. Then we will evaluate if the tuned model performs better than the base model.

- In the end, we will choose 1 model as our best model which then will be used for our user.

```python
# Modeling
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, fbeta_score, plot_roc_curve
from sklearn.model_selection import cross_val_score
```

```python
# Logistic Regression

lr_fe = LogisticRegression()

lr_fe.fit(X_train_fe_prep, y_train)

print('Baseline Logistic Regression: ')
y_pred = lr_fe.predict(X_train_fe_prep)
print('Performance on train set')
print('Accuracy: ', accuracy_score(y_train, y_pred))
print('Precision: ', precision_score(y_train, y_pred))
print('Recall: ', recall_score(y_train, y_pred))
print('F1: ', f1_score(y_train, y_pred))
print('F2: ', fbeta_score(y_train, y_pred, beta=2))
print('\n')
print('Mean Accuracy CV: ',cross_val_score(lr_fe, X_train_fe_prep, y_train).mean())
print('Mean Precision CV: ', cross_val_score(lr_fe, X_train_fe_prep, y_train, scoring='precision').mean())
print('\n')
print('Performance on test set')
y_pred_test = lr_fe.predict(X_test_fe_prep)
print('Accuracy: ', accuracy_score(y_test, y_pred_test))
print('Precision: ', precision_score(y_test, y_pred_test))
print('Recall: ', recall_score(y_test, y_pred_test))
print('F1: ', f1_score(y_test, y_pred_test))
print('F2: ', fbeta_score(y_test, y_pred_test, beta=2))
```

```python
# KNN

knn = KNeighborsClassifier()
knn.fit(X_train_fe_prep, y_train)
print('Baseline KNN: ')
y_pred = knn.predict(X_train_fe_prep)
print('Performance on train set')
print('Accuracy: ', accuracy_score(y_train, y_pred))
print('Precision: ', precision_score(y_train, y_pred))
print('Recall: ', recall_score(y_train, y_pred))
print('F1: ', f1_score(y_train, y_pred))
print('F2: ', fbeta_score(y_train, y_pred, beta=2))
print('\n')
print('Mean Accuracy CV: ',cross_val_score(knn, X_train_fe_prep, y_train).mean())
print('Mean Precision CV: ', cross_val_score(knn, X_train_fe_prep, y_train, scoring='precision').mean())
print('\n')
print('Performance on test set')
y_pred_test = knn.predict(X_test_fe_prep)
print('Accuracy: ', accuracy_score(y_test, y_pred_test))
print('Precision: ', precision_score(y_test, y_pred_test))
print('Recall: ', recall_score(y_test, y_pred_test))
print('F1: ', f1_score(y_test, y_pred_test))
print('F2: ', fbeta_score(y_test, y_pred_test, beta=2))
```

```python
# Random Forest

rf_fe = RandomForestClassifier(random_state=22)
rf_fe.fit(X_train_fe_prep_tree, y_train)
print('Baseline Random Forest: ')
y_pred = rf_fe.predict(X_train_fe_prep_tree)
print('Performance on train set')
print('Accuracy: ', accuracy_score(y_train, y_pred))
print('Precision: ', precision_score(y_train, y_pred))
print('Recall: ', recall_score(y_train, y_pred))
print('F1: ', f1_score(y_train, y_pred))
print('F2: ', fbeta_score(y_train, y_pred, beta=2))
print('\n')
print('Mean Accuracy CV: ',cross_val_score(rf_fe, X_train_fe_prep_tree, y_train).mean())
print('Mean Precision CV: ', cross_val_score(rf_fe, X_train_fe_prep_tree, y_train, scoring='precision').mean())
print('\n')
print('Performance on test set')
y_pred_test = rf_fe.predict(X_test_fe_prep_tree)
print('Accuracy: ', accuracy_score(y_test, y_pred_test))
print('Precision: ', precision_score(y_test, y_pred_test))
print('Recall: ', recall_score(y_test, y_pred_test))
print('F1: ', f1_score(y_test, y_pred_test))
print('F2: ', fbeta_score(y_test, y_pred_test, beta=2))
```

```python

# XGBoost
xgb_fe = XGBClassifier(random_state=22, verbosity=0)
xgb_fe.fit(X_train_fe_prep_tree, y_train)

print('Baseline XGB: ')
y_pred = xgb_fe.predict(X_train_fe_prep_tree)
print('Performance on train set')
print('Accuracy: ', accuracy_score(y_train, y_pred))
print('Precision: ', precision_score(y_train, y_pred))
print('Recall: ', recall_score(y_train, y_pred))
print('F1: ', f1_score(y_train, y_pred))
print('F2: ', fbeta_score(y_train, y_pred, beta=2))
print('\n')
print('Mean Accuracy CV: ',cross_val_score(xgb_fe, X_train_fe_prep_tree, y_train).mean())
print('Mean Precision CV: ', cross_val_score(xgb_fe, X_train_fe_prep_tree, y_train, scoring='precision').mean())
print('\n')
print('Performance on test set')
y_pred_test = xgb_fe.predict(X_test_fe_prep_tree)
print('Accuracy: ', accuracy_score(y_test, y_pred_test))
print('Precision: ', precision_score(y_test, y_pred_test))
print('Recall: ', recall_score(y_test, y_pred_test))
print('F1: ', f1_score(y_test, y_pred_test))
print('F2: ', fbeta_score(y_test, y_pred_test, beta=2))
```

# **Model Selection & Conclusion**
 From the results of the train and test models of the four tested models, it can be said that the XGBoost model is the one that can be used with an accuracy and precision value in the test set of 86% for both and shows a stable value between the train set and the test set. So **XGBoost** is the chosen model.
